```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("msarmi9/food101tiny")

print("Path to dataset files:", path)
```

```python
data_path = os.path.join(path, "data","food-101-tiny", "train", "apple_pie")
```

```python
data_dir = data_path

image_files = os.listdir(data_dir)
print(f"Number of images in {data_path}: {len(image_files)}")
print("First 5 images:", image_files[:5])
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
first_image_path = os.path.join(data_dir, image_files[0])
image = mpimg.imread(first_image_path)
print("Image shape:", image.shape) 
```

```python
plt.figure(figsize=(10, 5))
for i in range(5):
    img_path = os.path.join(data_dir, image_files[i])
    img = mpimg.imread(img_path)
    plt.subplot(1, 5, i+1)
    plt.imshow(img)
    plt.axis("off")
plt.show()
```

```python
image_normalized = image / 255.0
print("Pixel values range (before normalization):", image.min(), "-", image.max())
print("Pixel values range (after normalization):", image_normalized.min(), "-", image_normalized.max())
```

```python
import numpy as np
mean = np.mean(image_normalized)
std = np.std(image_normalized)
print(f"Mean pixel value: {mean:.4f}")
print(f"Std pixel value: {std:.4f}")
```
